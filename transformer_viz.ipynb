```python
import numpy as np
import matplotlib.pyplot as plt
import torch

from torchtransformers.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
```

```python
t = np.linspace(0, 2 * np.pi, 100)
y = np.sin(t) + 3
plt.plot(t, y)
plt.show()
```

```python
ypt = torch.tensor(y.reshape(-1, 1))

plt.plot(t, MinMaxScaler().fit_transform(ypt))
plt.plot(t, (y - np.min(y)) / np.ptp(y))
plt.show()

plt.plot(t, StandardScaler().fit_transform(ypt))
plt.plot(t, (y - y.mean()) / y.std(ddof=1))
plt.show()

plt.plot(t, RobustScaler().fit_transform(ypt))
plt.plot(t, (y - np.median(y)) / np.diff(np.quantile(y, [0.25, 0.75])))
plt.show()
```
